```python
import matplotlib.pyplot as plt
```

```python
# define a function to do the math on a given starting number
def collatz(n):
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence
```

```python
# Test the function with a starting number
starting_number = 95
sequence = collatz(starting_number)

# Plotting the sequence
plt.plot(sequence)#, marker='x')      # marker is the type of point used in graph
plt.title('Collatz Conjecture Sequence')
plt.xlabel('Step')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
